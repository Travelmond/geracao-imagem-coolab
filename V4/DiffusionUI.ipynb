# Célula 0: Enviar cache-manager para o Google Drive
Faça upload do arquivo ZIP da extensão cache-manager. O notebook extrai no Drive automaticamente.

```python
from google.colab import files
import zipfile, os, shutil

print('📦 CÉLULA 0: ENVIANDO cache-manager PARA O DRIVE')
print('═' * 55)
print()
print('📋 Como preparar o ZIP:')
print('   1. No seu computador, entre na pasta do projeto')
print('   2. Clique com botão direito na pasta cache-manager/')
print('   3. Selecione "Comprimir" ou "Enviar para → Pasta compactada"')
print('   4. O arquivo se chamará cache-manager.zip')
print()
print('👇 Clique no botão "Escolher arquivos" abaixo e selecione o .zip:')
print()

try:
    uploaded = files.upload()
    
    zip_found = None
    for name, data in uploaded.items():
        if name.endswith('.zip'):
            zip_found = (name, data)
            break
    
    if zip_found is None:
        print('❌ Nenhum arquivo .zip encontrado!')
        print('   Certifique-se de selecionar cache-manager.zip')
    else:
        zip_name, zip_data = zip_found
        print(f'📦 {zip_name} recebido ({len(zip_data)/1024:.1f} KB)')
        print()
        
        dest = '/content/drive/MyDrive/Stable_Diffusion_Dados/cache-manager'
        
        if os.path.exists(dest):
            print('🗑️ Removendo cache-manager antigo...')
            shutil.rmtree(dest)
        
        os.makedirs(dest, exist_ok=True)
        
        zip_path = f'/tmp/{zip_name}'
        with open(zip_path, 'wb') as f:
            f.write(zip_data)
        
        print('📂 Extraindo arquivos...')
        with zipfile.ZipFile(zip_path, 'r') as zf:
            zf.extractall(dest)
        
        inner = os.path.join(dest, 'cache-manager')
        if os.path.isdir(inner):
            for item in os.listdir(inner):
                src = os.path.join(inner, item)
                dst = os.path.join(dest, item)
                if os.path.exists(dst):
                    if os.path.isdir(dst):
                        shutil.rmtree(dst)
                    else:
                        os.remove(dst)
                shutil.move(src, dst)
            shutil.rmtree(inner)
            print('   📁 Estrutura corrigida (subpasta removida)')
        
        os.remove(zip_path)
        
        print()
        print('🔍 Verificando arquivos essenciais:')
        required = [
            'install.py',
            'scripts/tab_ui.py',
            'scripts/cache_manager.py',
            'scripts/session_manager.py',
            'scripts/hardware_detector.py',
            'scripts/memory_monitor.py',
            'scripts/forge_bridge.py',
        ]
        all_ok = True
        for f in required:
            path = os.path.join(dest, f)
            if os.path.exists(path):
                print(f'   ✅ {f}')
            else:
                print(f'   ❌ {f} — NÃO encontrado!')
                all_ok = False
        
        if all_ok:
            print()
            print('✅ cache-manager instalado no Drive com sucesso!')
            print(f'   📁 {dest}')
            print('   👉 Execute a Célula 1 agora.')
        else:
            print()
            print('⚠️ Alguns arquivos não foram encontrados.')
            print('   Verifique se o ZIP contém a estrutura correta.')
            print('   Estrutura esperada: cache-manager/install.py, scripts/...')

except Exception as e:
    print(f'❌ Erro: {e}')
    print('   Tente novamente ou coloque a pasta manualmente no Drive.')
```

# Célula 3: Vacinas e Dependências
Corrige incompatibilidades de bibliotecas.

```python
# @title 🚑 Célula 3: Diagnóstico, Cura e Preparação do Ambiente (Híbrido + Manifesto)
import os, sys, time, subprocess, shutil, json
from pathlib import Path
from datetime import datetime
from enum import Enum
from dataclasses import dataclass, asdict
from typing import List, Optional, Dict

class Status(Enum):
    HEALTHY = "healthy"
    FIXED = "fixed"
    FAILED = "failed"
    SKIPPED = "skipped"

@dataclass
class VaccineResult:
    name: str
    status: Status
    message: str
    duration: float
    version: Optional[str] = None

@dataclass
class EnvironmentManifest:
    timestamp: str
    python_version: str
    torch_version: str
    cuda_version: str
    libraries: Dict[str, str]
    total_vaccines: int
    successful: int
    failed: int
    warnings: List[str]

class EnvironmentDoctor:
    def __init__(self):
        self.python = "/usr/bin/python3.10"
        self.pip = f"{self.python} -m pip"
        self.wheels_dir = "/content/drive/MyDrive/Stable_Diffusion_Dados/wheels"
        self.lock_file = "/content/env_ready.lock"
        self.manifest_file = "/content/env_manifest.json"
        self.constraints_file = "/content/pip_constraints_sd.txt"
        self.results: List[VaccineResult] = []
        self.warnings: List[str] = []
        
        # Versões universais testadas e aprovadas
        self.UNIVERSAL_VERSIONS = {
            "numpy": "1.26.4",
            "scipy": "1.13.1",
            "opencv-python-headless": "4.10.0.84",
            "onnxruntime-gpu": "1.17.1",
            "onnx": "1.16.1",
            "insightface": "0.7.3",
            "albumentations": "1.4.3",
            "bitsandbytes": "0.43.3",
            "protobuf": "3.20.3",
            "scikit-image": "0.21.0",
            "pillow": "9.5.0"
        }
        
        os.makedirs(self.wheels_dir, exist_ok=True)

    def run_command(self, cmd: str, desc: str = "Executando", timeout: int = 300, quiet: bool = False) -> tuple:
        if not quiet:
            print(f"⏳ {desc}...")
        try:
            result = subprocess.run(cmd, shell=True, capture_output=True, text=True, timeout=timeout, env={**os.environ, "PIP_NO_INPUT": "1"})
            if result.returncode != 0:
                return False, result.stderr[:500] if result.stderr else "Unknown error"
            return True, result.stdout[:500] if result.stdout else ""
        except subprocess.TimeoutExpired:
            return False, f"Timeout after {timeout}s"
        except Exception as e:
            return False, str(e)

    def check_import(self, lib_name: str, version_constraint: str = None) -> bool:
        cmd = f"{self.python} -c 'import {lib_name}; print({lib_name}.__version__ if hasattr({lib_name}, "__version__") else "unknown")'"
        ok, output = self.run_command(cmd, f"Verificando {lib_name}", timeout=10, quiet=True)
        if not ok:
            return False
        if version_constraint and version_constraint in output:
            return True
        elif version_constraint:
            return False
        return ok

    def get_installed_version(self, lib_name: str) -> Optional[str]:
        cmd = f"{self.pip} show {lib_name} | grep Version"
        ok, output = self.run_command(cmd, "Obtendo versão", timeout=10, quiet=True)
        if ok and "Version:" in output:
            return output.split("Version:")[1].strip()
        return None

    def save_wheel_to_drive(self, lib_name: str):
        cache_dir = "/root/.cache/pip/wheels"
        if os.path.exists(cache_dir):
            for root, dirs, files in os.walk(cache_dir):
                for file in files:
                    if file.startswith(lib_name.replace("-", "_")) and file.endswith(".whl"):
                        src = os.path.join(root, file)
                        dst = os.path.join(self.wheels_dir, file)
                        if not os.path.exists(dst):
                            shutil.copy2(src, dst)
                            print(f"💾 Wheel salvo: {file}")
                        return

    def install_with_strategies(self, lib_name: str, strategies: List[str], version: str = None) -> bool:
        for i, cmd in enumerate(strategies):
            ok, err = self.run_command(cmd, f"Tentativa {i+1} para {lib_name}", timeout=300)
            if ok and self.check_import(lib_name.split("==")[0] if "==" in lib_name else lib_name):
                self.save_wheel_to_drive(lib_name.split("==")[0])
                return True
            time.sleep(2)
        self.warnings.append(f"Falha ao instalar {lib_name}: {err[:100]}")
        return False

    def install_numpy(self) -> VaccineResult:
        start = time.time()
        version = self.UNIVERSAL_VERSIONS["numpy"]
        if self.check_import("numpy", version):
            return VaccineResult("NumPy", Status.HEALTHY, f"Versão {version} já instalada", time.time()-start, version)
        self.run_command(f"{self.pip} uninstall numpy -y", "Removendo NumPy antigo", quiet=True)
        strategies = [f"{self.pip} install numpy=={version} --quiet", f"{self.pip} install numpy=={version} --no-cache-dir --quiet"]
        success = self.install_with_strategies(f"numpy=={version}", strategies, version)
        if success:
            installed_version = self.get_installed_version("numpy")
            return VaccineResult("NumPy", Status.FIXED, f"Instalado {installed_version}", time.time()-start, installed_version)
        else:
            return VaccineResult("NumPy", Status.FAILED, "Todas tentativas falharam", time.time()-start)

    def install_clip(self) -> VaccineResult:
        start = time.time()
        if self.check_import("clip"):
            version = self.get_installed_version("clip") or "unknown"
            return VaccineResult("CLIP", Status.HEALTHY, "Já instalado", time.time()-start, version)
        strategies = [
            f"test -f {self.wheels_dir}/clip*.whl && {self.pip} install --no-deps {self.wheels_dir}/clip*.whl --quiet",
            f"unset PIP_CONSTRAINT && {self.pip} install --no-build-isolation 'https://github.com/openai/CLIP/archive/d50d76daa670286dd6cacf3bcd80b5e4823fc8e1.zip' --quiet",
            f"unset PIP_CONSTRAINT && {self.pip} install git+https://github.com/openai/CLIP.git --quiet"
        ]
        success = self.install_with_strategies("clip", strategies)
        if success:
            return VaccineResult("CLIP", Status.FIXED, "Instalado com sucesso", time.time()-start)
        else:
            return VaccineResult("CLIP", Status.FAILED, "Falha na instalação", time.time()-start)

    def install_bitsandbytes(self) -> VaccineResult:
        start = time.time()
        version = self.UNIVERSAL_VERSIONS["bitsandbytes"]
        if self.check_import("bitsandbytes"):
            return VaccineResult("bitsandbytes", Status.HEALTHY, "Já instalado", time.time()-start)
        strategies = [
            f"test -f {self.wheels_dir}/bitsandbytes*.whl && {self.pip} install --no-deps {self.wheels_dir}/bitsandbytes*.whl --quiet",
            f"unset PIP_CONSTRAINT && {self.pip} install bitsandbytes=={version} --quiet",
            f"unset PIP_CONSTRAINT && {self.pip} install bitsandbytes-cuda121 --quiet"
        ]
        success = self.install_with_strategies("bitsandbytes", strategies)
        if success:
            return VaccineResult("bitsandbytes", Status.FIXED, "Instalado", time.time()-start)
        else:
            return VaccineResult("bitsandbytes", Status.FAILED, "Falha na instalação", time.time()-start)

    def install_onnx_runtime(self) -> VaccineResult:
        start = time.time()
        onnx_version = self.UNIVERSAL_VERSIONS["onnx"]
        onnxruntime_version = self.UNIVERSAL_VERSIONS["onnxruntime-gpu"]
        if self.check_import("onnxruntime"):
            return VaccineResult("ONNX Runtime", Status.HEALTHY, "Já instalado", time.time()-start)
        self.run_command(f"{self.pip} install onnx=={onnx_version} --quiet", "Instalando ONNX")
        strategies = [
            f"test -f {self.wheels_dir}/onnxruntime_gpu*.whl && {self.pip} install --no-deps {self.wheels_dir}/onnxruntime_gpu*.whl --quiet",
            f"unset PIP_CONSTRAINT && {self.pip} install onnxruntime-gpu=={onnxruntime_version} --extra-index-url https://aiinfra.pkgs.visualstudio.com/PublicPackages/_packaging/onnxruntime-cuda-12/pypi/simple/ --quiet",
            f"unset PIP_CONSTRAINT && {self.pip} install onnxruntime --quiet"
        ]
        success = self.install_with_strategies("onnxruntime", strategies)
        if success:
            return VaccineResult("ONNX Runtime", Status.FIXED, "Instalado", time.time()-start)
        else:
            return VaccineResult("ONNX Runtime", Status.FAILED, "Falha na instalação", time.time()-start)

    def install_insightface(self) -> VaccineResult:
        start = time.time()
        version = self.UNIVERSAL_VERSIONS["insightface"]
        if self.check_import("insightface"):
            return VaccineResult("InsightFace", Status.HEALTHY, "Já instalado", time.time()-start)
        self.run_command(f"{self.pip} install onnx=={self.UNIVERSAL_VERSIONS['onnx']} --quiet", "Preparando ONNX")
        self.run_command(f"{self.pip} install albumentations=={self.UNIVERSAL_VERSIONS['albumentations']} --quiet", "Preparando Albumentations")
        strategies = [
            f"test -f {self.wheels_dir}/insightface*.whl && {self.pip} install --no-deps {self.wheels_dir}/insightface*.whl --quiet",
            f"unset PIP_CONSTRAINT && {self.pip} install insightface=={version} onnx=={self.UNIVERSAL_VERSIONS['onnx']} --quiet",
            f"unset PIP_CONSTRAINT && {self.pip} install insightface --quiet"
        ]
        success = self.install_with_strategies("insightface", strategies)
        if success:
            return VaccineResult("InsightFace", Status.FIXED, "Instalado", time.time()-start)
        else:
            return VaccineResult("InsightFace", Status.FAILED, "Falha na instalação", time.time()-start)

    def setup_constraints(self):
        constraints = "\n".join([f"{lib}=={ver}" for lib, ver in self.UNIVERSAL_VERSIONS.items()])
        with open(self.constraints_file, 'w') as f:
            f.write(constraints)
        os.environ['PIP_CONSTRAINT'] = self.constraints_file
        print(f"🛡️ PIP_CONSTRAINT definido: {len(self.UNIVERSAL_VERSIONS)} versões travadas")

    def create_manifest(self):
        manifest = EnvironmentManifest(
            timestamp=datetime.now().isoformat(),
            python_version=subprocess.check_output([self.python, "--version"], text=True, stderr=subprocess.STDOUT).strip().split()[1],
            torch_version=self.get_installed_version("torch") or "unknown",
            cuda_version=subprocess.check_output("nvcc --version 2>/dev/null | grep release || echo 'unknown'", shell=True, text=True).strip(),
            libraries={lib: self.get_installed_version(lib) or "not_installed" for lib in self.UNIVERSAL_VERSIONS.keys()},
            total_vaccines=len(self.results),
            successful=sum(1 for r in self.results if r.status in [Status.HEALTHY, Status.FIXED]),
            failed=sum(1 for r in self.results if r.status == Status.FAILED),
            warnings=self.warnings
        )
        with open(self.manifest_file, 'w') as f:
            json.dump(asdict(manifest), f, indent=2)
        print(f"📋 Manifesto salvo: {self.manifest_file}")
        return manifest

    def create_lock(self):
        with open(self.lock_file, 'w') as f:
            f.write(f"Ready at {datetime.now().isoformat()}")
        print(f"🔒 Lock criado: {self.lock_file}")

    def heal(self):
        print("="*60)
        print("🚀 INICIANDO AUTO-HEALING HÍBRIDO COM MANIFESTO")
        print("="*60)
        self.setup_constraints()
        vaccines = [("NumPy", self.install_numpy), ("CLIP", self.install_clip), ("bitsandbytes", self.install_bitsandbytes), ("ONNX Runtime", self.install_onnx_runtime), ("InsightFace", self.install_insightface)]
        for name, func in vaccines:
            result = func()
            self.results.append(result)
            emoji = "✅" if result.status == Status.HEALTHY else "🔧" if result.status == Status.FIXED else "❌"
            print(f"{emoji} {name}: {result.status.value} ({result.duration:.1f}s)")
            if result.version:
                print(f"   └─ Versão: {result.version}")
        self.create_manifest()
        self.create_lock()
        successful = sum(1 for r in self.results if r.status in [Status.HEALTHY, Status.FIXED])
        failed = sum(1 for r in self.results if r.status == Status.FAILED)
        print("\n" + "="*60)
        print(f"📊 RESUMO: {successful}/{len(vaccines)} bem-sucedidas, {failed} falhas")
        if self.warnings:
            print(f"⚠️ Avisos: {len(self.warnings)}")
            for w in self.warnings[:3]:
                print(f"   - {w[:80]}...")
        print("="*60)

if __name__ == "__main__":
    doctor = EnvironmentDoctor()
    doctor.heal()
```

# Célula 4: Instalar Extensões
Instala o cache-manager e extensões do Forge.

```python
import os, shutil

print('🔌 ETAPA 4: INSTALANDO EXTENSÕES')
print('─' * 50)

drive_path = '/content/drive/MyDrive/Stable_Diffusion_Dados'
forge_dir = '/content/stable-diffusion-webui-forge'
ext_path = f'{forge_dir}/extensions'
os.makedirs(ext_path, exist_ok=True)

print('\n📦 cache-manager:')
cache_manager_src = f'{drive_path}/cache-manager'
cache_manager_dst = f'{ext_path}/cache-manager'

try:
    if not os.path.exists(cache_manager_src):
        print(f'   ❌ NÃO encontrado em: {cache_manager_src}')
        print('   💡 Copie a pasta cache-manager/ do repositório para o Drive!')
    else:
        if os.path.exists(cache_manager_dst):
            shutil.rmtree(cache_manager_dst)
        shutil.copytree(cache_manager_src, cache_manager_dst)
        
        tab_ui = f'{cache_manager_dst}/scripts/tab_ui.py'
        install = f'{cache_manager_dst}/install.py'
        if os.path.exists(tab_ui) and os.path.exists(install):
            print('   ✅ cache-manager copiado com sucesso!')
        else:
            missing = []
            if not os.path.exists(tab_ui): missing.append('scripts/tab_ui.py')
            if not os.path.exists(install): missing.append('install.py')
            print(f'   ⚠️ Arquivos faltando: {missing}')
except Exception as e:
    print(f'   ❌ Erro ao copiar cache-manager: {e}')

reactor_src = f'{drive_path}/sd-webui-reactor'
reactor_dst = f'{ext_path}/sd-webui-reactor'

if os.path.exists(reactor_src):
    print('\n📦 sd-webui-reactor (Face Swap):')
    try:
        if os.path.exists(reactor_dst):
            shutil.rmtree(reactor_dst)
        shutil.copytree(reactor_src, reactor_dst)
        if os.path.exists(f'{reactor_dst}/install.py'):
            print('   ✅ ReActor copiado com sucesso!')
        else:
            print('   ⚠️ ReActor: install.py não encontrado')
    except Exception as e:
        print(f'   ❌ Erro ao copiar ReActor: {e}')
else:
    print('\nℹ️ sd-webui-reactor não encontrado no Drive.')
    print('   Execute a Célula 0.5 se quiser face swap (opcional).')

print('\n📦 Extensões adicionais:')
extensions = [
    ('sd-webui-civbrowser', 'https://github.com/SignalFlagZ/sd-webui-civbrowser.git'),
]

for name, url in extensions:
    dst = f'{ext_path}/{name}'
    try:
        if not os.path.exists(dst):
            !git clone {url} {dst}
            if os.path.exists(dst):
                print(f'   ✅ {name}')
            else:
                print(f'   ❌ {name} — clone falhou')
        else:
            print(f'   ℹ️ {name} — já instalado')
    except Exception as e:
        print(f'   ❌ {name}: {e}')

print('\n✅ ETAPA 4 CONCLUÍDA!')
```

# Célula 4.5: Download de Modelos por URL
Baixa modelos do Civitai, HuggingFace ou outras fontes diretamente para o Google Drive.

```python
import os, json, subprocess

print('📥 DOWNLOAD DE MODELOS POR URL')
print('═' * 50)

drive_path = '/content/drive/MyDrive/Stable_Diffusion_Dados'

chave = ''
config_path = f'{drive_path}/.cache_config.json'
try:
    if os.path.exists(config_path):
        with open(config_path) as f:
            chave = json.load(f).get('civitai_api_key', '')
            if chave:
                print(f'🔑 API Key carregada ({len(chave)} caracteres)')
except Exception as e:
    print(f'⚠️ Erro ao ler config: {e}')

tipo_map = {
    'checkpoint': 'Modelos_Base',
    'lora': 'LoRAs',
    'vae': 'VAEs',
    'text_encoder': 'Text_Encoders'
}

print('\nTipos disponíveis: checkpoint, lora, vae, text_encoder')
tipo = input('Tipo do modelo: ').strip().lower()

if tipo not in tipo_map:
    print(f'❌ Tipo inválido: "{tipo}". Use um dos tipos acima.')
else:
    url = input('URL do download: ').strip()
    nome = input('Nome do arquivo (ex: modelo.safetensors): ').strip()

    if not url or not nome:
        print('❌ URL ou nome vazio. Abortado.')
    else:
        pasta = tipo_map[tipo]
        destino = f'{drive_path}/{pasta}/{nome}'

        if not os.path.exists(f'{drive_path}/{pasta}'):
            print(f'❌ Pasta de destino não existe: {drive_path}/{pasta}')
            print('   Execute a Célula 1 primeiro.')
        else:
            try:
                if chave and 'civitai' in url.lower():
                    sep = '&' if '?' in url else '?'
                    url = f'{url}{sep}token={chave}'
                    print('🔑 API Key injetada na URL')

                print(f'⏳ Baixando {nome}...')
                result = subprocess.run(
                    ['wget', '-q', '--show-progress', '-O', destino, url],
                    timeout=1800
                )

                if result.returncode == 0 and os.path.exists(destino) and os.path.getsize(destino) > 50000:
                    size_mb = os.path.getsize(destino) / (1024 ** 2)
                    size_str = f'{size_mb:.1f} MB' if size_mb < 1024 else f'{size_mb/1024:.1f} GB'
                    print(f'✅ {nome} baixado! ({size_str})')
                    print('💡 Use a aba Cache Manager no Forge para acelerar o acesso.')
                else:
                    print(f'❌ Download falhou (exit code: {result.returncode})')
                    print('   Verifique a URL e sua conexão.')
                    if os.path.exists(destino):
                        os.remove(destino)
            except subprocess.TimeoutExpired:
                print(f'❌ Timeout (30 min). Arquivo muito grande ou conexão lenta.')
                if os.path.exists(destino):
                    os.remove(destino)
            except Exception as e:
                print(f'❌ Erro: {e}')
                if os.path.exists(destino):
                    os.remove(destino)
```

# Célula 4.6: Upload de Modelos do PC
Envia arquivos do seu computador diretamente para o Google Drive.

```python
from google.colab import files

print('📤 UPLOAD DE MODELOS DO PC')
print('═' * 50)

tipo_map = {
    'checkpoint': 'Modelos_Base',
    'lora': 'LoRAs',
    'vae': 'VAEs',
    'text_encoder': 'Text_Encoders'
}

print('Tipos: checkpoint, lora, vae, text_encoder')
tipo = input('Tipo do modelo: ').strip().lower()

if tipo not in tipo_map:
    print(f'❌ Tipo inválido: "{tipo}". Use um dos tipos acima.')
else:
    pasta = tipo_map[tipo]
    destino = f'/content/drive/MyDrive/Stable_Diffusion_Dados/{pasta}'

    if not os.path.exists(destino):
        print(f'❌ Pasta destino não existe: {destino}')
        print('   Execute a Célula 1 primeiro.')
    else:
        print(f'📁 Salvos em: {destino}')
        print('👇 Selecione os arquivos no popup abaixo:')

        try:
            arquivos = files.upload()

            if not arquivos:
                print('⚠️ Nenhum arquivo selecionado. Operação cancelada.')
            else:
                sucessos = 0
                falhas = 0
                for nome, conteudo in arquivos.items():
                    try:
                        caminho = f'{destino}/{nome}'
                        with open(caminho, 'wb') as f:
                            f.write(conteudo)
                        size_mb = len(conteudo) / (1024**2)
                        size_str = f'{size_mb:.1f} MB' if size_mb < 1024 else f'{size_mb/1024:.1f} GB'
                        print(f'   ✅ {nome} ({size_str})')
                        if os.path.exists(nome):
                            os.remove(nome)
                        sucessos += 1
                    except Exception as e:
                        print(f'   ❌ {nome}: {e}')
                        falhas += 1

                print(f'\n✅ Upload: {sucessos} sucesso(s), {falhas} falha(s)')
                print('💡 Use a aba Cache Manager no Forge para acelerar o acesso.')
        except Exception as e:
            print(f'❌ Erro no upload: {e}')
```

# Célula 5: Verificação Final
Verifica se tudo está configurado corretamente.

```python
import os

print('🔍 ETAPA 5: VERIFICAÇÃO FINAL')
print('═' * 50)

drive_path = '/content/drive/MyDrive/Stable_Diffusion_Dados'
forge_dir = '/content/stable-diffusion-webui-forge'
checks = []

checks.append(('Google Drive montado', os.path.ismount('/content/drive')))

for folder in ['Modelos_Base', 'LoRAs', 'VAEs', 'Text_Encoders', 'Imagens_Geradas', 'logs']:
    checks.append((f'Drive/{folder}', os.path.isdir(f'{drive_path}/{folder}')))

for d in ['checkpoints', 'loras', 'vaes', 'text_encoders']:
    checks.append((f'Cache: /content/cache/{d}', os.path.isdir(f'/content/cache/{d}')))

symlink_checks = {
    f'{forge_dir}/models/Stable-diffusion': '/content/cache/checkpoints',
    f'{forge_dir}/models/Lora': '/content/cache/loras',
    f'{forge_dir}/models/VAE': '/content/cache/vaes',
    f'{forge_dir}/models/text_encoder': '/content/cache/text_encoders',
    f'{forge_dir}/outputs': '/content/outputs_temp'
}
for link, expected_target in symlink_checks.items():
    try:
        is_ok = os.path.islink(link) and os.readlink(link) == expected_target and os.path.exists(expected_target)
        checks.append((f'Symlink: {os.path.basename(link)}', is_ok))
    except Exception:
        checks.append((f'Symlink: {os.path.basename(link)}', False))

ext_cache = f'{forge_dir}/extensions/cache-manager'
ext_tab = f'{ext_cache}/scripts/tab_ui.py'
ext_install = f'{ext_cache}/install.py'
checks.append(('Extensão cache-manager', os.path.isdir(ext_cache) and os.path.isfile(ext_tab) and os.path.isfile(ext_install)))

checks.append(('Forge launch.py', os.path.exists(f'{forge_dir}/launch.py')))

try:
    result = !python3.10 --version
    version = result.s if hasattr(result, 's') else str(result)
    checks.append(('Python 3.10', '3.10' in version))
except Exception:
    checks.append(('Python 3.10', False))

print('\n  CHECKLIST DE CONFIGURAÇÃO')
print('═' * 50)
all_ok = True
for name, status in checks:
    icon = '✅' if status else '❌'
    print(f'  {icon} {name}')
    if not status:
        all_ok = False

print('═' * 50)
if all_ok:
    print('  ✅ TUDO CONFIGURADO!')
    print()
    print('  👉 Acesse a aba "💾 Cache Manager" no Forge')
    print('     para gerenciar modelos entre Drive e cache.')
else:
    print('  ⚠️ Alguns itens falharam.')
    print()
    print('  Verifique os ❌ acima e execute as células anteriores novamente.')
    print('  Dica: Execute as células em ordem (1→2→3→4→4.2→5)')
print('═' * 50)
```

# Célula 0: Enviar cache-manager para o Google Drive
Faça upload do arquivo ZIP da extensão cache-manager. O notebook extrai no Drive automaticamente.

```python
from google.colab import files
import zipfile, os, shutil

print('📦 CÉLULA 0: ENVIANDO cache-manager PARA O DRIVE')
print('═' * 55)
print()
print('📋 Como preparar o ZIP:')
print('   1. No seu computador, entre na pasta do projeto')
print('   2. Clique com botão direito na pasta cache-manager/')
print('   3. Selecione "Comprimir" ou "Enviar para → Pasta compactada"')
print('   4. O arquivo se chamará cache-manager.zip')
print()
print('👇 Clique no botão "Escolher arquivos" abaixo e selecione o .zip:')
print()

try:
    uploaded = files.upload()
    
    zip_found = None
    for name, data in uploaded.items():
        if name.endswith('.zip'):
            zip_found = (name, data)
            break
    
    if zip_found is None:
        print('❌ Nenhum arquivo .zip encontrado!')
        print('   Certifique-se de selecionar cache-manager.zip')
    else:
        zip_name, zip_data = zip_found
        print(f'📦 {zip_name} recebido ({len(zip_data)/1024:.1f} KB)')
        print()
        
        dest = '/content/drive/MyDrive/Stable_Diffusion_Dados/cache-manager'
        
        # Limpa destino se existir
        if os.path.exists(dest):
            print('🗑️ Removendo cache-manager antigo...')
            shutil.rmtree(dest)
        
        os.makedirs(dest, exist_ok=True)
        
        # Extrai ZIP
        zip_path = f'/tmp/{zip_name}'
        with open(zip_path, 'wb') as f:
            f.write(zip_data)
        
        print('📂 Extraindo arquivos...')
        with zipfile.ZipFile(zip_path, 'r') as zf:
            zf.extractall(dest)
        
        # Se extraiu dentro de subpasta, move para o nível certo
        inner = os.path.join(dest, 'cache-manager')
        if os.path.isdir(inner):
            for item in os.listdir(inner):
                src = os.path.join(inner, item)
                dst = os.path.join(dest, item)
                if os.path.exists(dst):
                    if os.path.isdir(dst):
                        shutil.rmtree(dst)
                    else:
                        os.remove(dst)
                shutil.move(src, dst)
            shutil.rmtree(inner)
            print('   📁 Estrutura corrigida (subpasta removida)')
        
        os.remove(zip_path)
        
        # Verificação
        print()
        print('🔍 Verificando arquivos essenciais:')
        required = [
            'install.py',
            'scripts/tab_ui.py',
            'scripts/cache_manager.py',
            'scripts/session_manager.py',
            'scripts/hardware_detector.py',
            'scripts/memory_monitor.py',
            'scripts/forge_bridge.py',
        ]
        all_ok = True
        for f in required:
            path = os.path.join(dest, f)
            if os.path.exists(path):
                print(f'   ✅ {f}')
            else:
                print(f'   ❌ {f} — NÃO encontrado!')
                all_ok = False
        
        if all_ok:
            print()
            print('✅ cache-manager instalado no Drive com sucesso!')
            print(f'   📁 {dest}')
            print('   👉 Execute a Célula 1 agora.')
        else:
            print()
            print('⚠️ Alguns arquivos não foram encontrados.')
            print('   Verifique se o ZIP contém a estrutura correta.')
            print('   Estrutura esperada: cache-manager/install.py, scripts/...')

except Exception as e:
    print(f'❌ Erro: {e}')
    print('   Tente novamente ou coloque a pasta manualmente no Drive.')
```

# Célula 4.5: Download de Modelos por URL
Baixa modelos do Civitai, HuggingFace ou outras fontes diretamente para o Google Drive.

```python
import os, json, subprocess

print('📥 DOWNLOAD DE MODELOS POR URL')
print('═' * 50)

drive_path = '/content/drive/MyDrive/Stable_Diffusion_Dados'

# Carregar API Key do config
chave = ''
config_path = f'{drive_path}/.cache_config.json'
try:
    if os.path.exists(config_path):
        with open(config_path) as f:
            chave = json.load(f).get('civitai_api_key', '')
            if chave:
                print(f'🔑 API Key carregada ({len(chave)} caracteres)')
except Exception as e:
    print(f'⚠️ Erro ao ler config: {e}')

tipo_map = {
    'checkpoint': 'Modelos_Base',
    'lora': 'LoRAs',
    'vae': 'VAEs',
    'text_encoder': 'Text_Encoders'
}

print('\nTipos disponíveis: checkpoint, lora, vae, text_encoder')
tipo = input('Tipo do modelo: ').strip().lower()

if tipo not in tipo_map:
    print(f'❌ Tipo inválido: "{tipo}". Use um dos tipos acima.')
else:
    url = input('URL do download: ').strip()
    nome = input('Nome do arquivo (ex: modelo.safetensors): ').strip()

    if not url or not nome:
        print('❌ URL ou nome vazio. Abortado.')
    else:
        pasta = tipo_map[tipo]
        destino = f'{drive_path}/{pasta}/{nome}'

        # Verificar se pasta destino existe
        if not os.path.exists(f'{drive_path}/{pasta}'):
            print(f'❌ Pasta de destino não existe: {drive_path}/{pasta}')
            print('   Execute a Célula 1 primeiro.')
        else:
            try:
                # Injetar token Civitai
                if chave and 'civitai' in url.lower():
                    sep = '&' if '?' in url else '?'
                    url = f'{url}{sep}token={chave}'
                    print('🔑 API Key injetada na URL')

                print(f'⏳ Baixando {nome}...')
                result = subprocess.run(
                    ['wget', '-q', '--show-progress', '-O', destino, url],
                    timeout=1800
                )

                if result.returncode == 0 and os.path.exists(destino) and os.path.getsize(destino) > 50000:
                    size_mb = os.path.getsize(destino) / (1024 ** 2)
                    size_str = f'{size_mb:.1f} MB' if size_mb < 1024 else f'{size_mb/1024:.1f} GB'
                    print(f'✅ {nome} baixado! ({size_str})')
                    print('💡 Use a aba Cache Manager no Forge para acelerar o acesso.')
                else:
                    print(f'❌ Download falhou (exit code: {result.returncode})')
                    print('   Verifique a URL e sua conexão.')
                    if os.path.exists(destino):
                        os.remove(destino)
            except subprocess.TimeoutExpired:
                print(f'❌ Timeout (30 min). Arquivo muito grande ou conexão lenta.')
                if os.path.exists(destino):
                    os.remove(destino)
            except Exception as e:
                print(f'❌ Erro: {e}')
                if os.path.exists(destino):
                    os.remove(destino)
```

# Célula 4.6: Upload de Modelos do PC

# Célula 5: Verificação Final
Verifica se tudo está configurado corretamente.

```python
import os

print('🔍 ETAPA 5: VERIFICAÇÃO FINAL')
print('═' * 50)

drive_path = '/content/drive/MyDrive/Stable_Diffusion_Dados'
forge_dir = '/content/stable-diffusion-webui-forge'
checks = []

# Drive montado
checks.append(('Google Drive montado', os.path.ismount('/content/drive')))

# Pastas Drive
for folder in ['Modelos_Base', 'LoRAs', 'VAEs', 'Text_Encoders', 'Imagens_Geradas', 'logs']:
    checks.append((f'Drive/{folder}', os.path.isdir(f'{drive_path}/{folder}')))

# Cache local
for d in ['checkpoints', 'loras', 'vaes', 'text_encoders']:
    checks.append((f'Cache: /content/cache/{d}', os.path.isdir(f'/content/cache/{d}')))

# Symlinks (verificação real)
symlink_checks = {
    f'{forge_dir}/models/Stable-diffusion': '/content/cache/checkpoints',
    f'{forge_dir}/models/Lora': '/content/cache/loras',
    f'{forge_dir}/models/VAE': '/content/cache/vaes',
    f'{forge_dir}/models/text_encoder': '/content/cache/text_encoders',
    f'{forge_dir}/outputs': '/content/outputs_temp'
}
for link, expected_target in symlink_checks.items():
    try:
        is_ok = os.path.islink(link) and os.readlink(link) == expected_target and os.path.exists(expected_target)
        checks.append((f'Symlink: {os.path.basename(link)}', is_ok))
    except Exception:
        checks.append((f'Symlink: {os.path.basename(link)}', False))

# Extensão cache-manager (verificação profunda)
ext_cache = f'{forge_dir}/extensions/cache-manager'
ext_tab = f'{ext_cache}/scripts/tab_ui.py'
ext_install = f'{ext_cache}/install.py'
checks.append(('Extensão cache-manager', os.path.isdir(ext_cache) and os.path.isfile(ext_tab) and os.path.isfile(ext_install)))

# Forge
checks.append(('Forge launch.py', os.path.exists(f'{forge_dir}/launch.py')))

# Python 3.10
try:
    result = !python3.10 --version 2>&1
    version = result.s if hasattr(result, 's') else str(result)
    checks.append(('Python 3.10', '3.10' in version))
except Exception:
    checks.append(('Python 3.10', False))

# Imprimir checklist
print('\n  CHECKLIST DE CONFIGURAÇÃO')
print('═' * 50)
all_ok = True
for name, status in checks:
    icon = '✅' if status else '❌'
    print(f'  {icon} {name}')
    if not status:
        all_ok = False

print('═' * 50)
if all_ok:
    print('  ✅ TUDO CONFIGURADO!')
    print()
    print('  👉 Acesse a aba "💾 Cache Manager" no Forge')
    print('     para gerenciar modelos entre Drive e cache.')
else:
    print('  ⚠️ Alguns itens falharam.')
    print()
    print('  Verifique os ❌ acima e execute as células anteriores novamente.')
    print('  Dica: Execute as células em ordem (1→2→3→4→5)')
print('═' * 50)
```
